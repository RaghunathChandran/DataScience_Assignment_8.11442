```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import sqlite3 as db
```

```python
#from pandasql import sqldf
adultDf = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", skipinitialspace=True)
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
sqladb = adultDf.copy()
headers = ['age', 'workclass','fnlwgt', 'education','educationnum', 'maritalstatus','occupation','relationship','race','sex',
          'capitalgain','capitalloss', 'hoursperweek','nativecountry', 'lessThanEqualTo50K']
sqladb.columns= headers
```

#### 1. Select 10 records from the adult sqladb

```python
pysqldf("SELECT * FROM sqladb LIMIT 10;")
```

#### 2. Show me the average hours per week of all men who are working in private sector `

```python
avgworkinghoursQuery = '''SELECT 
                            AVG(hoursperweek) AS AverageHoursPerWeek 
                            FROM sqladb WHERE workclass ="Private" '''

pysqldf(avgworkinghoursQuery)
```

#### 3. Show me the frequency table for education, occupation and relationship, separately 

```python
educationfrequencyTable = '''SELECT education 'Education', 
                    COUNT (education) 'Frequency'
                    FROM sqladb
                    GROUP BY education'''

pysqldf(educationfrequencyTable)
```

```python
occupationfrequencyTable = '''SELECT occupation 'Occupation', count (occupation) 'Frequency'
                    FROM sqladb
                    WHERE occupation != "?"
                    GROUP BY occupation'''

pysqldf(occupationfrequencyTable)
```

```python
relationshipfrequencyTable = '''SELECT relationship 'Relationship', 
                                COUNT (relationship) 'Frequency'
                                FROM sqladb
                                GROUP BY relationship'''

pysqldf(relationshipfrequencyTable)
```

#### 4. Are there any people who are married, working in private sector and having a masters degree

```python
marriedMastersQuery = '''SELECT * FROM sqladb 
                            WHERE workclass = "Private" 
                            AND maritalstatus NOT IN ("Divorced","Separated","Widowed", "Never-married")
                            AND education = "Masters"
                            '''
pysqldf(marriedMastersQuery)
```

#### 5. What is the average, minimum and maximum age group for people working in different sectors 

```python
avgminMaxAgeBySector = '''SELECT 
                occupation,
                AVG(age) AS AverageAge,
                MIN(age) AS MinimumAge,
                MAX(age) AS MaximumAge               
                FROM sqladb 
                WHERE occupation != "?"
                GROUP BY occupation'''
pysqldf(avgminMaxAgeBySector)
```

#### 6. Calculate age distribution by country 

```python
ageDistributionByCountry = '''SELECT 
                nativecountry AS Country,
                COUNT(nativecountry) AS Count,             
                age
                FROM sqladb 
                where nativecountry != "?"
                GROUP BY nativecountry, age'''
pysqldf(ageDistributionByCountry)
```

#### 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss' 

```python
netCapitalGainQuery = '''SELECT age, workclass, education, maritalstatus, 
                        occupation, sex, capitalgain, capitalloss,
                        (capitalgain-capitalloss) AS NetCapitalGain
                        FROM sqladb'''
pysqldf(netCapitalGainQuery)
```
